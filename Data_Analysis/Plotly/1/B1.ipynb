```python
# Импортируем необходимые библиотеки
import plotly.graph_objects as go
import numpy as np
import pandas as pd
#  Следующие библиотеки необходимы вам, чтобы скачать dataset
import io
import requests
```

```python
# Читаем данные из датасета — мы используем https://github.com/jbrownlee/Datasets/blob/master/airline-passengers.csv
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
s = requests.get(url).content
data = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

Возможности Plotly простираются куда дальше, чем просто включение и выключение графиков из легенды. Вы можете добавлять различные элементы User Interface, используя которые, пользователь может безгранично взаимодействовать с вашей инфографикой.

Здесь предложен пример графика по всё тем же данным авиаперевозок. Выделим основные элементы:
- Слайдер — позволяет указать год, за который мы хотим просматривать данные.
- Три кнопки, которые позволяют переключать режим отображения: линейный график, пирожковая и столбчатые диаграммы.
- Выпадающий список, или drop down, — который также позволяет переключать режим отображения.

Получилось, кажется, отлично, теперь попробуем сделать такой график вместе.

```python
paper = go.Figure()

for year in years:
  x = data.loc[data['year'] == year, 'Month_name']
  y = data.loc[data['year'] == year, 'Passengers']
  paper.add_trace(
      go.Scatter(
          x = x,
          y = y,
          visible = False,
          name = year
      )
  )

  steps = []

  for i in range(len(paper.data)):
    step = dict(
        method='update',
        args = [
            {'visible': [t == i for t in range(len(paper.data))]},
            {'title.text': f'Airlaine Passengers for {paper.data[i].name} Year'}
        ]
    )
    steps.append(step)
```

```python
# paper.data
# steps
```

```python
sliders = [dict(
    currentvalue = {'prefix': 'Year'},
    steps = steps
)]

start_index = 0
paper.data[start_index].visible = True

```

```python
paper.update_layout(
    sliders=sliders,
    title={'text': 'Airline Passengers for {} Year'.format(paper.data[start_index].name)},
    updatemenus=[
        dict(
            type = 'buttons',
            direction = 'left',
            buttons = list([
                dict(
                    args = ['type', 'scatter'],
                    label='Lines mode',
                    method = 'restyle'),

                dict(
                    args = ['type', 'pie'],
                    labels = 'Pie mode',
                    method = 'restyle'),

                dict(args = ['type', 'bar'],
                     labels = 'Bar mode',
                     method = 'restyle')
            ]),

            showactive = True,
            x = 0.0,
            xanchor = 'left',
            y = 1.125,
            yanchor = 'top'
        ),

        dict(
            buttons = list([
                dict(
                    args = ['type', 'scatter'],
                    label = 'Lines mode',
                    method = 'restyle'),
                dict(
                    args = ['type', 'pie'],
                    label = 'Pie mode',
                    method = 'restyle'),
                dict(
                    args = ['type', 'bar'],
                    labels = 'Bar mode',
                    method = 'restyle')
                ]),
            y = 0.9,
            xanchor = 'right'
        )
    ]
)


# paper['layout']['sliders'][0]['currentvalue']['prefix'] = 'Year: '
```

```python
paper['layout']['sliders'][0]['currentvalue']['prefix']='Year: '
for i, date in enumerate(years):
    date_str = str(date)
    paper['layout']['sliders'][0]['steps'][i]['label']=date_str

paper.show()
```
