Даны точки $x$ и значения в этих точках $y$. Нормализуйте $y$ с помощью *z-score* и постройте график зависимости нормализованных значений от $x$. Для графика используйте `.scatter(x,y_tansformed)`

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt

x = np.linspace(1,10,num=10)
y = np.array(
    [1.,  3.,  4.,  2., 10.,  5.,  5.,  2.,  5., 10.],
    dtype=np.float32
)

z_scaled_data = StandardScaler().fit_transform(y.reshape(-1, 1)).reshape(-1)
```

```python
plt.scatter(x, z_scaled_data)
plt.title('Correlations beetwen feature and z-scaled target data')
plt.xlabel('feature')
plt.ylabel('Z-scaled target')
plt.show()
```
